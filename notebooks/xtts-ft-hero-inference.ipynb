```python
from IPython.display import Audio, display
from TTS.api import TTS
from TTS.utils.manage import ModelManager
from TTS.tts.models.xtts import Xtts
from TTS.tts.configs.xtts_config import XttsConfig

from tqdm.notebook import tqdm
import librosa
```

```python
ModelManager._download_model_files(
    ["https://coqui.gateway.scarf.sh/hf-coqui/XTTS-v2/main/vocab.json"],
    "/kaggle/working/",
    progress_bar=False
)
```

```python
CONFIG_PATH = "/kaggle/input/xtts-hero-gothic/pytorch/default/3/xtts-hero-model/config.json"
XTTS_CHECKPOINT = "/kaggle/input/xtts-hero-gothic/pytorch/default/3/xtts-hero-model/best_model.pth"
TOKENIZER_PATH = "/kaggle/working/vocab.json"
SPEAKER_REFERENCE = "/kaggle/input/ljspeech-gothic/15/wavs/DIA_DUSTY_METMELVIN_15_02.wav"
```

```python
config = XttsConfig()
config.load_json(CONFIG_PATH)
model = Xtts.init_from_config(config)
model.load_checkpoint(config, checkpoint_path=XTTS_CHECKPOINT, vocab_path=TOKENIZER_PATH, use_deepspeed=False)
model.cuda()
None
```

```python
gpt_cond_latent, speaker_embedding = model.get_conditioning_latents(audio_path=[SPEAKER_REFERENCE])
```

```python
# Info_GorHanis_ChargeGood_15_01
out = model.inference(
    text="Да, я готов.",
    language="ru",
    gpt_cond_latent=gpt_cond_latent,
    speaker_embedding=speaker_embedding,
    temperature=0.6,
    speed=1.3,
    enable_text_splitting=True
)
Audio(librosa.resample(out["wav"], orig_sr=24000, target_sr=22050), rate=22050)
```
